```python
plt.title("Nigeria 2023 Election Participation by State")
plt.xlabel("Turnout Percentage(%)")
plt.ylabel("States")
x_ticks = np.arange(0, 50, 2)
y_ticks = np.arange(0, 150, 5)
plt.yticks(x_ticks)
plt.xticks(y_ticks)
plt.margins(0.02)
#plt.subplots_adjust(bottom=0.15)
plt.grid(color="lightgray", alpha=0.5, zorder=1)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.bar(x, y, color="lightgreen", alpha=0.9, zorder=2, width=0.68)

plt.show()
```

```python
plt.title("Politically Active States(PAS)")
plt.xlabel("Turnout Percentage(%)")
plt.ylabel("States")
x_ticks = np.arange(0, 59, 1)
plt.xticks(x_ticks, rotation=-60)
plt.margins(0.002)
#plt.subplots_adjust(bottom=0.15)
plt.grid(color="lightgray", alpha=0.5, zorder=1)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.bar(x[27:], y[27:], color="red", alpha=0.9, zorder=2, width=0.68)

plt.show()
```

```python
plt.title("Least Politically Active States(LAS)")
plt.xlabel("Turnout Percentage(%)")
plt.ylabel("States")
x_ticks = np.arange(0, 59, 1)
plt.xticks(x_ticks, rotation=-60)
plt.margins(0.02)
plt.margins(0.02)
#plt.subplots_adjust(bottom=0.15)
plt.grid(color="lightgray", alpha=0.5, zorder=1)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.bar(x[0:10], y[0:10], color="lightblue", alpha=0.9, zorder=2, width=0.68)

plt.show()
```
